```python
df['Placed']=np.where(df['Placed']=='Yes',1,0)
```

```python
# Verify column names
#print(df.columns)

# Drop columns (handling missing columns)
df = df.drop(['Name', 'Student_ID', 'Gender', 'Department', 'Salary (LPA)', 'Extracurricular_Activities'], axis=1, errors='ignore')

# Compute correlation
correlation = df.corr()
correlation
```

```python
import seaborn as sns
sns.heatmap(correlation)
```

```python
new_df=df[['CGPA','IQ','Internships','Placed']]
new_df
```

```python
from sklearn.model_selection import train_test_split
x=new_df.drop('Placed',axis=1)
y=new_df['Placed']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(18,10))
plt.plot(y_test,"mo",label='Actual')
plt.plot(y_pred,"go",label='Predict')

plt.title("Comparison of Actual vs Predicted Values", fontsize=14)
plt.xlabel("Sample Index", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)

plt.show()
```

```python
# prompt: GRPAH B/W PLACED OR NOT PLACED B OF ACTUAL AND TRAINED DATA  BOTH GRPAH FACED TO EACH OTHER

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot actual values
axes[0].plot(y_test, label='Actual')
axes[0].set_title("Actual Placement Data", fontsize=14)
axes[0].set_xlabel("Sample Index", fontsize=12)
axes[0].set_ylabel("Placed (1=Yes, 0=No)", fontsize=12)
axes[0].legend(fontsize=12)
axes[0].grid(True)

# Plot predicted values
axes[1].plot(y_pred, "g", label='Predicted')
axes[1].set_title("Predicted Placement Data", fontsize=14)
axes[1].set_xlabel("Sample Index", fontsize=12)
axes[1].set_ylabel("Placed (1=Yes, 0=No)", fontsize=12)
axes[1].legend(fontsize=12)
axes[1].grid(True)

plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()
```
